```python
import pandas as pd
```

```python
df = pd.read_csv('ITIL 22-10-2021.csv')
```

```python
df.head()
```

```python
for i in range(len(df)):
    if (df["Site Type"][i] == "SAP Team") and (df["Sub-category"][i] == "Reset Password "):
        df["Sub-category"] = df["Sub-category"].replace(["Reset Password "], "Reset Password SAP")
```

```python
df.to_csv('ITIL 25-10-2021-cleaned.csv', index = False, header=True)
```
