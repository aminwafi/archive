```python
import pandas as pd
```

```python
df = pd.read_csv('Threat Logs OCT.csv')
```

```python
df.head()
```

```python
df['Action']
```

```python
df.dtypes
```

```python
df_2 = pd.DataFrame(df["Action"], columns=['A', 'B'])
```

```python
df_2.head()
```

```python
df_2["A"] = df["Action"]
```

```python
df_2.head()
```

```python
df_2["B"][df_2['A'].str.contains("quarantined", na=False)] = "Quarantined"
df_2["B"][df_2['A'].str.contains("Quarantine", na=False)] = "Quarantined"
df_2["B"][df_2['A'].str.contains("cleaned", na=False)] = "Cleaned"
df_2["B"][df_2['A'].str.contains("Unable", na=False)] = "Unsuccessful upload file"
df_2["B"][df_2['A'].str.contains("Access denied", na=False)] = "Access denied"
df_2["B"][df_2['A'].str.contains("Block", na=False)] = "Block"
```

```python
df_2.head(20)
```

```python
df_final = pd.concat([df, df_2], axis=1)
```

```python
df_final.head(20)
```

```python
df_final = df_final.drop(columns=['A'])
```

```python
df_final.to_csv('Threat Logs OCT.csv', index = False, header = True)
```
