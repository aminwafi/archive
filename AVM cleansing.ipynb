```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AVM Cloud.csv")
df.head()
```

```python
df.ID
```

```python
list(df.ID)
```

```python
df_2 = pd.DataFrame(df["Description"], columns=['A', 'B'])
```

```python
df_2.head()
```

```python
df_2['A'] = df["Description"]
```

```python
df_2["B"][df_2['A'].str.contains("CPU", na=False)] = "CPU"
df_2["B"][df_2['A'].str.contains("Memory", na=False)] = "Memory"
df_2["B"][df_2['A'].str.contains("Disk", na=False)] = "Disk"
df_2["B"][df_2['A'].str.contains("Server", na=False)] = "Server"
df_2["B"][df_2['A'].str.contains("Resource Host", na=False)] = "Server"
df_2["B"][df_2['A'].str.contains("SAP QAS", na=False)] = "Server"
df_2["B"][df_2['A'].str.contains("Critical Vulnerability", na=False)] = "Server"
df_2["B"][df_2['A'].str.contains("DR", na=False)] = "DR"
df_2["B"][df_2['A'].str.contains("Backup", na=False)] = "Backup"
df_2["B"][df_2['A'].str.contains("backup", na=False)] = "Backup"
df_2["B"][df_2['A'].str.contains("VPN", na=False)] = "VPN"
df_2["B"][df_2['A'].str.contains("VM", na=False)] = "VM"
```

```python
df_2.head(20)
```

```python
df_2["A"] = df_2["A"].fillna('')
```

```python
for i, row in df_2.iterrows():
    if 'VM' in row["A"] and 'Memory' in row["A"]:
        row["B"] = 'VM & Memory'
    elif 'VM' in row["A"] and 'Memory' in row["A"] and 'CPU' in row["A"]:
        row["B"] = 'VM & Memory & CPU'
    elif 'CPU' in row["A"] and 'Memory' in row["A"]:
        row["B"] = 'CPU & Memory'
```

```python
df_2.head(20)
```

```python
df_2["B"] = df_2["B"].fillna('Others')
```

```python
df_2.head(20)
```

```python
df_final = pd.concat([df, df_2], axis=1)
```

```python
df_final.head(30)
```

```python
df_final = df_final.drop(columns=['A', 'Sub-category'])
```

```python
df_final.head()
```

```python
df_final = df_final.rename(columns={'B': 'Sub-Category'})
```

```python
df_final.head()
```

```python
df_final.to_csv('AVM Ticketing Report.csv', index = False, header=True)
```
