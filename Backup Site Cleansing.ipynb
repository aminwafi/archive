```python
import pandas as pd
```

```python
df = pd.read_csv('LPP Backup History (Daily).csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Start Date/Time'] = pd.to_datetime(df['Start Date/Time'])
df['End Date/Time'] = pd.to_datetime(df['End Date/Time'])
```

```python
df['Start Date/Time'] = df['Start Date/Time'].dt.strftime('%d/%m/%y %H:%M:%S')
df['End Date/Time'] = df['End Date/Time'].dt.strftime('%d/%m/%y %H:%M:%S')
```

```python
df.head()
```

```python
df.to_csv("Backup LPP.csv", index=False, header=True)
```
