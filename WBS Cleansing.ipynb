```python
import pandas as pd
```

```python
df = pd.read_csv("Renewable Energy.csv")
```

```python
df.head()
```

```python
df = df.dropna(axis=1, how="all")
```

```python
df = df.dropna()
```

```python
df.head(100)
```

```python
df_2 = pd.DataFrame(df["People"], columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'Use-Case', 'Outcome'])
```

```python
df_2.head()
```

```python
df_2["A"] = df["People"]
df_2["B"] = df["Process"]
df_2["C"] = df["Technology"]
df_2["D"] = df["Experience"]
df_2["E"] = df["Drive Decision"]
df_2["F"] = df["Process Efficiency"]
df_2["G"] = df["Scalability"]
```

```python
df_2.head()
```

```python
df_2["A"] = df_2["A"].astype(str)
df_2["B"] = df_2["B"].astype(str)
df_2["C"] = df_2["C"].astype(str)
df_2["D"] = df_2["D"].astype(str)
df_2["E"] = df_2["E"].astype(str)
df_2["F"] = df_2["F"].astype(str)
df_2["G"] = df_2["G"].astype(str)
```

```python
for i, row in df_2.iterrows():
    if '1' in row["A"] and '1' in row["B"] and '1' in row["C"]:
        row["Use-Case"] = 'People,Process,Technology'
    elif '1' in row["A"] and '1' in row["B"]:
        row["Use-Case"] = 'People,Process'
    elif '1' in row["A"] and '1' in row["C"]:
        row["Use-Case"] = 'People,Technology'
    elif '1' in row["B"] and '1' in row["C"]:
        row["Use-Case"] = "Process,Technology"
```

```python
for i, row in df_2.iterrows():
    if '1' in row["D"] and '1' in row["E"] and '1' in row["F"] and '1' in row['G']:
        row["Outcome"] = 'Drive Overall Experience,Drive Decisions,Drive Process Efficiency,Drive Scalability'
    elif '1' in row["D"] and '1' in row["E"] and '1' in row["F"]:
        row["Outcome"] = 'Drive Overall Experience,Drive Decisions,Drive Process Efficiency'
    elif '1' in row["D"] and '1' in row["E"] and '1' in row["G"]:
        row["Outcome"] = 'Drive Overall Experience,Drive Decisions,Drive Scalability'
    elif '1' in row["D"] and '1' in row["E"]:
        row["Outcome"] = 'Drive Overall Experience,Drive Decisions'
    elif '1' in row["D"] and '1' in row["F"]:
        row["Outcome"] = 'Drive Overall Experience,Drive Process Efficiency'
    elif '1' in row["D"] and '1' in row["G"]:
        row["Outcome"] = 'Drive Overall Experience,Drive Scalability'
    elif '1' in row["E"] and '1' in row["F"] and '1' in row["G"]:
        row["Outcome"] = 'Drive Decisions,Drive Process Efficiency,Drive Scalability'
    elif '1' in row["E"] and '1' in row["F"]:
        row["Outcome"] = 'Drive Decisions,Drive Process Efficiency'
    elif '1' in row["E"] and '1' in row["G"]:
        row["Outcome"] = 'Drive Decisions,Drive Scalability'
    elif '1' in row["F"] and '1' in row["G"]:
        row["Outcome"] = 'Drive Process Efficiency,Drive Scalability'
```

```python
df_2["Use-Case"] = df_2["Use-Case"].fillna('Test')
df_2["Outcome"] = df_2["Outcome"].fillna('Test')
```

```python
df_2.head(20)
```

```python
for i, row in df_2.iterrows():
    if '1' in row["A"] and 'Test' in row["Use-Case"]:
        row['Use-Case'] = 'People'
    elif '1' in row["B"] and 'Test' in row["Use-Case"]:
        row['Use-Case'] = 'Process'
    elif '1' in row["C"] and 'Test' in row["Use-Case"]:
        row['Use-Case'] = 'Technology'
```

```python
for i, row in df_2.iterrows():
    if '1' in row["D"] and 'Test' in row["Outcome"]:
        row['Outcome'] = 'Drive Overall Experience'
    elif '1' in row["E"] and 'Test' in row["Outcome"]:
        row['Outcome'] = 'Drive Decisions'
    elif '1' in row["F"] and 'Test' in row["Outcome"]:
        row["Outcome"] = 'Drive Process Efficiency'
    elif '1' in row["G"] and 'Test' in row["Outcome"]:
        row["Outcome"] = 'Drive Scalability'
```

```python
df_2.head(50)
```

```python
df_final = pd.concat([df, df_2], axis=1)
```

```python
df_final.head(50)
```

```python
df_final = df_final.drop(columns=['A', 'B', 'C', 'D', 'E', 'F', 'G'])
```

```python
df_final.head(50)
```

```python
df_final['Use-Case'] = df_final['Use-Case'].str.split(',')
```

```python
df_final = df_final.explode('Use-Case')
```

```python
df_final.head(50)
```

```python
df_final['Outcome'] = df_final['Outcome'].str.split(',')
```

```python
df_final.head()
```

```python
df_final = df_final.explode('Outcome')
```

```python
df_final.head(50)
```

```python
nan_value = float("NaN")
df_final['Use-Case'][df_final['Use-Case'].str.contains('Test', na=False)] = nan_value
df_final['Outcome'][df_final['Outcome'].str.contains('Test', na=False)] = nan_value
```

```python
df_final.head(100000)
```

```python
df_final = df_final.dropna()
```

```python
df_final.to_csv('RE.csv', index = False, header=True)
```
