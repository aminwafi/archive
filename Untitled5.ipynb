```python
import pandas as pd
```

```python
df = pd.read_csv('Threat Logs.csv')
```

```python
df.head()
```

```python
df['Action']
```

```python
df.dtypes
```

```python
df['Threat Result'] = df['Threat Result'].fillna('ba')
```

```python
df.head()
```

```python
for i in range(len(df)):
    if (df["Action"][i] == "File quarantined"):
        df["Threat Result"] = df["Threat Result"].replace(["ba"], "Quarantined")
    elif (df["Action"][i] == "File cleaned"):
        df["Threat Result"] = df["Threat Result"].replace(["ba"], "Cleaned")
    elif (df["Action"][i] == "Access denied"):
        df["Threat Result"] = df["Threat Result"].replace(["ba"], "Access Denied")
    elif (df["Action"][i] == "Unable to upload file"):
        df["Threat Result"] = df["Threat Result"].replace(["ba"], "Unsuccessful")
    else:
        df["Threat Result"] = df["Threat Result"].replace(["ba"], "")
```

```python
df.head()
```

```python
df.to_csv('Threat Logs-cleaned.csv', index = False, header = True)
```
