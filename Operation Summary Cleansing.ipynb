```python
df_new = pd.read_csv('Virtual Machine Performance (June - Sept).csv')
```

```python
df_new.head()
```

```python
df_new['Memory Max Usage']
```

```python
df_3 = pd.DataFrame(df_new["Name"], columns=['A', 'B', 'C', 'D', 'CPU Average Usage Indicator', 'CPU Max Usage Indicator', 'Memory Average Usage Indicator', 'Memory Max Usage Indicator'])
```

```python
df_3.head()
```

```python
df_3['A'] = df_new['CPU Average Usage']
df_3['B'] = df_new['CPU Max Usage']
df_3['C'] = df_new['Memory Average Usage']
df_3['D'] = df_new['Memory Max Usage']
```

```python
df_3.loc[(df_3['A'] >= 80.00) & (df_3['A'] <= 89.99), 'CPU Average Usage Indicator'] = 'Warning'
df_3.loc[(df_3['A'] >= 90.00) & (df_3['A'] <= 100.00), 'CPU Average Usage Indicator'] = 'Critical'
df_3.loc[(df_3['B'] >= 80.00) & (df_3['B'] <= 89.99), 'CPU Max Usage Indicator'] = 'Warning'
df_3.loc[(df_3['B'] >= 90.00) & (df_3['B'] <= 100.00), 'CPU Max Usage Indicator'] = 'Critical'
df_3.loc[(df_3['C'] >= 80.00) & (df_3['C'] <= 89.99), 'Memory Average Usage Indicator'] = 'Warning'
df_3.loc[(df_3['C'] >= 90.00) & (df_3['C'] <= 100.00), 'Memory Average Usage Indicator'] = 'Critical'
df_3.loc[(df_3['D'] >= 80.00) & (df_3['D'] <= 89.99), 'Memory Max Usage Indicator'] = 'Warning'
df_3.loc[(df_3['D'] >= 90.00) & (df_3['D'] <= 100.00), 'Memory Max Usage Indicator'] = 'Critical'
```

```python
df_3.head()
```

```python
df_3['A'] = df_3['A'].fillna('-')
df_3['B'] = df_3['B'].fillna('-')
df_3['C'] = df_3['C'].fillna('-')
df_3['D'] = df_3['D'].fillna('-')
```

```python
df_3.head(1000)
```

```python
df_3.fillna('')
```

```python
df_3['A'] = df_3['A'].astype(str)
df_3['B'] = df_3['B'].astype(str)
df_3['C'] = df_3['C'].astype(str)
df_3['D'] = df_3['D'].astype(str)
df_3['CPU Average Usage Indicator'] = df_3['CPU Average Usage Indicator'].astype(str)
df_3['CPU Max Usage Indicator'] = df_3['CPU Max Usage Indicator'].astype(str)
df_3['Memory Average Usage Indicator'] = df_3['Memory Average Usage Indicator'].astype(str)
df_3['Memory Max Usage Indicator'] = df_3['Memory Max Usage Indicator'].astype(str)
```

```python
for i, row in df_3.iterrows():
    if (not('-' in row['A'])) and not('Critical' in row['CPU Average Usage Indicator']) and not ('Warning' in row['CPU Average Usage Indicator']):
        row['CPU Average Usage Indicator'] = 'Safe'

for i, row in df_3.iterrows():
    if (not('-' in row['B'])) and not('Critical' in row['CPU Max Usage Indicator']) and not ('Warning' in row['CPU Max Usage Indicator']):
        row['CPU Max Usage Indicator'] = 'Safe'

for i, row in df_3.iterrows():    
    if (not('-' in row['C'])) and not('Critical' in row['Memory Average Usage Indicator']) and not ('Warning' in row['Memory Average Usage Indicator']):
        row['Memory Average Usage Indicator'] = 'Safe'

for i, row in df_3.iterrows():    
    if (not('-' in row['D'])) and not('Critical' in row['Memory Max Usage Indicator']) and not ('Warning' in row['Memory Max Usage Indicator']):
        row['Memory Max Usage Indicator'] = 'Safe'
```

```python
df_3.head(1000)
```

```python
df_final2 = pd.concat([df_new, df_3], axis=1)
```

```python
df_final2.head()
```

```python
df_final2 = df_final2.drop(columns=['Indicator','A', 'B', 'C', 'D'])
```

```python
df_final2.head(1000)
```

```python
df_final2.to_csv('Virtual Machine Performance (June - Sept).csv')
```
