```python
df1 = pd.DataFrame()
```

```python
df1['Sub-category'] = df['Sub-category']
```

```python
for i, row in df1.iterrows():
    if 'Reset Password' in row['Sub-category']:
        row['Sub-category'] = 'Reset Password SAP'
```

```python
df = df.drop(columns=['Sub-category'])
```

```python
df['Created']
```

```python
df.dtypes
```

```python
df['Created'] = pd.to_datetime(df['Created'])
df['Resolved Date'] = pd.to_datetime(df['Resolved Date'])
df['Due Date'] = pd.to_datetime(df['Due Date'])
df['Close Date'] = pd.to_datetime(df['Close Date'])
```

```python
df['Created'] = df['Created'].dt.strftime('%d/%m/%Y %H:%M')
df['Resolved Date'] = df['Resolved Date'].dt.strftime('%d/%m/%Y %H:%M')
df['Due Date'] = df['Due Date'].dt.strftime('%d/%m/%Y %H:%M')
df['Close Date'] = df['Close Date'].dt.strftime('%d/%m/%Y %H:%M')
```
