```python
import pandas as pd
```

```python
df = pd.read_csv("ITIL.csv")
```

```python
df.head(10000)
```

```python
df_2 = pd.DataFrame(df["Issue Status"], columns=['A', 'Open Tickets'])
```

```python
df_2.head()
```

```python
df_2["A"] = df["Issue Status"]
```

```python
df_2.head()
```

```python
df_2["Open Tickets"][df_2['A'].str.contains("Open", na=False)] = '1'
```

```python
df_2["Open Tickets"] = df_2["Open Tickets"].fillna(0)
```

```python
df_2.head(30000)
```

```python
df_2 = df_2.drop(columns=['A'])
```

```python
df_final = pd.concat([df, df_2], axis=1)
```

```python
df_final.head()
```

```python
df_final.to_csv('ITIL.csv', index = False, header=True)
```
