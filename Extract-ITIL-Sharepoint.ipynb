```python
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium_stealth import stealth

import pandas as pd
import time
```

```python
driver = Chrome(executable_path = 'C:/Program Files (x86)/Google/Chrome/Application/chromedriver.exe')
driver.get('https://malakoffcommy.sharepoint.com/sites/OperationalDashboard/Shared%20Documents/Forms/AllItems.aspx?id=%2Fsites%2FOperationalDashboard%2FShared%20Documents%2FGeneral%2FOperational%20Data&viewid=47cd43ff%2D0341%2D4598%2Da22f%2Daa7f87d7b64e')
driver.maximize_window()

stealth(driver,
        languages=["en-US", "en"],
        vendor="Google Inc.",
        platform="Win32",
        webgl_vendor="Intel Inc.",
        renderer="Intel Iris OpenGL Engine",
        fix_hairline=True,
        )

chrome_options = Options()
chrome_options.add_argument("--headless")
```

```python
email = "amin1858@malakoff.com.my"
username = "amin1858"
password = "Malakoff@456!"

try:
    elem = WebDriverWait(driver, 100).until(
    EC.presence_of_element_located((By.NAME, "loginfmt"))
    )
except NoSuchElementException:
    time.sleep(2)

driver.find_element_by_name('loginfmt').send_keys(email)

time.sleep(2)

driver.find_element_by_id('idSIButton9').click()
```
