```python
import pandas as pd
```

```python
df = pd.read_csv('Threat Logs Nov.csv')
```

```python
df.head()
```

```python
df1 = pd.DataFrame(df["A"], columns=['Action'])
```

```python
df1["A"] = df["Action"]
```

```python
df1["A"][df1['A'].str.contains("quarantined", na=False)] = "Quarantined"
df1["A"][df1['A'].str.contains("Quarantine", na=False)] = "Quarantined"
df1["A"][df1['A'].str.contains("cleaned", na=False)] = "Cleaned"
df1["A"][df1['A'].str.contains("upload", na=False)] = "Upload File Denied"
df1["A"][df1['A'].str.contains("Access denied", na=False)] = "Access denied"
df1["A"][df1['A'].str.contains("Block", na=False)] = "Block"
```

```python
df1.head(51)
```

```python
df = df.drop(columns=['Action'])
```

```python
df.dtypes
```

```python
df1 = df1.rename({'A':'Action'}, axis='columns')
```

```python
df1.head()
```

```python
df_final = pd.concat([df, df1], axis=1)
```

```python
df_final.head()
```

```python
df_final['Generated'] = pd.to_datetime(df_final['Generated'])
```

```python
df_final['Received'] = pd.to_datetime(df_final['Received'])
```

```python
df_final['Generated'] = df_final['Generated'].dt.strftime('%d/%m/%y %H:%M:%S')
```

```python
df_final['Generated']
```

```python
df_final['Received'] = df_final['Received'].dt.strftime('%d/%m/%y %H:%M:%S')
```

```python
df_final['Received']
```

```python
df_final.to_csv('Threat Logs NOV.csv', index=False, header=True)
```
